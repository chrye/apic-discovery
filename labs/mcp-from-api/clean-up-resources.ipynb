### üóëÔ∏è Clean up resources

Delete the resource groups created by this demo. Set `IDX` to match the deployment index used in the demo notebook.

> ‚ö†Ô∏è Resource group deletion is **asynchronous** ‚Äî Azure will continue deleting in the background after the command returns. This is why the demo uses indexed names: you can start a new deployment with `IDX + 1` while the old one is still deleting.

```python
import os, sys
sys.path.insert(1, '../../shared')
import utils

# ‚îÄ‚îÄ Set the deployment index to clean up ‚îÄ‚îÄ
IDX = 2   # must match the IDX used in demo.ipynb

resource_groups = [
    f"rg-lab-mcp-containers-{IDX}",   # RG 3: FastMCP containers
    f"rg-lab-apic-central-{IDX}",     # RG 1: Central API Center
]

print(f"üóëÔ∏è  Deleting resource groups for deployment index {IDX}:")
for rg in resource_groups:
    output = utils.run(f"az group show --name {rg}", "", "")
    if output.success:
        print(f"  ‚Üí Deleting {rg} (async) ‚Ä¶")
        utils.run(f"az group delete --name {rg} --yes --no-wait", f"  ‚úÖ {rg} deletion started", f"  ‚ùå Failed to delete {rg}")
    else:
        print(f"  ‚è≠Ô∏è  {rg} ‚Äî does not exist, skipping")

print(f"\nüí° Deletion is async ‚Äî groups will disappear in ~2-5 minutes.")
print(f"   You can now redeploy with IDX = {IDX + 1} in demo.ipynb")
```
