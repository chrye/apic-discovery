### ğŸ—‘ï¸ Clean up resources

Delete the resource groups created by this demo (prod + staging + central). Set `IDX` to match the deployment index used in the demo notebook.

> âš ï¸ Resource group deletion is **asynchronous** â€” Azure will continue deleting in the background after the command returns. This is why the demo uses indexed names: you can start a new deployment with `IDX + 1` while the old one is still deleting.

```python
import os, sys
sys.path.insert(1, '../../shared')
import utils

# â”€â”€ Set the deployment index to clean up â”€â”€
IDX = 3   # must match the IDX used in demo.ipynb

ENVS = ["prod", "staging"]

# Build list of all resource groups (7 total: central + 3 per env Ã— 2 envs)
resource_groups = [f"rg-lab-apic-central-{IDX}"]  # RG 1: Central API Center
for env in ENVS:
    resource_groups.extend([
        f"rg-lab-mcp-demo-{env}-{IDX}",         # RG 2: RESTâ†’MCP APIs
        f"rg-lab-mcp-containers-{env}-{IDX}",    # RG 3: FastMCP containers
        f"rg-lab-a2a-demo-{env}-{IDX}",          # RG 4: A2A agents
    ])

print(f"ğŸ—‘ï¸  Deleting resource groups for deployment index {IDX} ({len(resource_groups)} groups):")
for rg in resource_groups:
    output = utils.run(f"az group show --name {rg}", "", "")
    if output.success:
        print(f"  â†’ Deleting {rg} (async) â€¦")
        utils.run(f"az group delete --name {rg} --yes --no-wait", f"  âœ… {rg} deletion started", f"  âŒ Failed to delete {rg}")
    else:
        print(f"  â­ï¸  {rg} â€” does not exist, skipping")

print(f"\nğŸ’¡ Deletion is async â€” groups will disappear in ~2-5 minutes.")
print(f"   You can now redeploy with IDX = {IDX + 1} in demo.ipynb")
```
